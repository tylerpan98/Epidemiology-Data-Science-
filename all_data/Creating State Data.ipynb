```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read in 2021 df
df2021 = pd.read_csv("Downloads/JH_2021.csv")
```

```python
#read in 2022 df
df2022 = pd.read_csv("Downloads/Testing_JH.csv")
```

```python
df2021.head()
```

```python
#append 2022 onto 2021 df
df=df2021.append([df2022], ignore_index=True)
```

```python
#remove time from date column
df['Last_Update'] = pd.to_datetime(df['Last_Update']).dt.date
```

```python
df.head()
```

```python
States = ['Wisconsin', 'Minnesota', 'Nebraska', 'Indiana', 'North Carolina', 'Kansas', 'Illinois', 'Texas', 'California', 'Iowa']
# keep top 10 highest agriculture producing states
df2=df[df.Province_State.isin(States) == True]
```

```python
df2.head()
```

```python
#sort by date column
df2= df2.sort_values(by=['Last_Update', 'Province_State'])
```

```python
#calculate the differences in total number of cases per day grouped by state, county
df2['cases_per_day'] = df2.groupby(['Province_State', 'Admin2'])['Confirmed'].diff()
```

```python
df2.to_csv('Downloads/JHbyCounty10States.csv')
```

```python
#sort by original df by date column
df= df.sort_values(by=['Last_Update', 'Province_State'])
```

```python
#sum number of confirmed cases and deaths by day by CDC Region
sum_df = df.groupby(['Province_State', 'Last_Update']).agg({'Confirmed': 'sum', 'Deaths': 'sum'})
```

```python
sum_df.head()
```

```python
#calculate the differences in total number of cases per day grouped by state, county
sum_df['cases_per_day'] = sum_df.groupby(['Province_State'])['Confirmed'].diff()
```

```python
sum_df.head()
```

```python
sum_df.to_csv('Downloads/JHbyState.csv')
```

```python
#read in JH Cases by state with negative values cleaned
dfstate = pd.read_csv("Downloads/JHbyStateNoNegs.csv")
```

```python
dfstate.head()
```

```python
#read in df with CDC region for each state
dfregion = pd.read_csv("Desktop/CDCRegion.csv")
```

```python
dfstate.describe
```

```python
dfstate['Date'] = pd.to_datetime(dfstate['Date'])
```

```python
#sum cases by weeks in CDC data (on saturdays)
dfstate = dfstate.groupby(['State', pd.Grouper(key='Date', freq='W-SAT')])['cases_per_day'].sum().reset_index()
```

```python
dfstate.head()
```

```python
dfstate['cases_per_day'].sum() 
```

```python
dfregion.head()
```

```python
#inner join on date to get weekly cumulative cases for the different variants
dfstate=pd.merge(dfstate,dfregion,left_on=['State'],right_on=['State Name'])
```

```python
dfstate.head()
```

```python
#read in df with CDC region variant proportion data for each region
dfvar = pd.read_csv("Downloads/CDCVariantProportionsDeDuped.csv")
```

```python
dfvar.head()
```

```python
#convert date column to datetime
dfvar['week_ending'] = pd.to_datetime(dfvar['week_ending'])
```

```python
#drop unnecessary columns
dfvar = dfvar.drop(['share_hi', 'share_lo', 'nchs_or_count_flag', 'modeltype'], axis=1)
```

```python
#inner join on region and date to merge DFs
dfjoin=pd.merge(dfstate,dfvar,left_on=['Date','CDC Region'] ,right_on=['week_ending', 'usa_or_hhsregion'])
```

```python
dfjoin.head()
```

```python
#multiply the variant share by the # new cases per week
dfjoin['var_cases_per_week'] = dfjoin.share * dfjoin.cases_per_day
```
